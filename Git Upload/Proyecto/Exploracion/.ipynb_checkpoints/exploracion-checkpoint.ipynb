## EXPLORACIÓN DE LAS RUTAS Y CARACTERÍSTICAS DE LOS DIFERENTES DATASETS 

```python
import pandas as pd

#Cargando los datos de entrenamiento 

train_data_1 = pd.read_xml(r"C:\Users\Ornella Gigante\OneDrive\Escritorio\training1.xml")
train_data_2 = pd.read_xml(r"C:\Users\Ornella Gigante\OneDrive\Escritorio\training2.xml")

```

```python
#Cargando los datos de prueba 

test_questions_data = pd.read_xml(r"C:\Users\Ornella Gigante\OneDrive\Escritorio\test1.xml")
test_data = pd.read_xml(r"C:\Users\Ornella Gigante\OneDrive\Escritorio\test2.xml")
qrels_data = pd.read_csv(r"C:\Users\Ornella Gigante\OneDrive\Escritorio\test3.txt", delimiter="\t", header=None)

```

## Limpieza de datos 

```python
#Reemplazo de datos Nan por "None" en datos de entrenamiento


train1_sin_nan = train_data_1.fillna("None")
train2_sin_nan = train_data_2.fillna("None")


# Mostrar el DataFrame sin valores NaN
print("=====================================")
print("Datos de prueba 1 sin Nan")
print("=====================================")
print(train1_sin_nan)
print("=====================================")
print("Datos de prueba 2 sin Nan")
print("=====================================")
print(train2_sin_nan)
```

```python
#Reemplazo de datos Nan por "None" en datos de prueba

test1_sin_nan = test_questions_data.fillna("None")
test2_sin_nan = test_data.fillna("None")
test3_sin_nan = qrels_data.fillna("None")


# Mostrar el DataFrame sin valores NaN
print("=====================================")
print("Datos de test 1 sin Nan")
print("=====================================")
print(test1_sin_nan)
print("=====================================")
print("Datos de test 2 sin Nan")
print("=====================================")
print(test2_sin_nan)
print("=====================================")
print("Datos de test 3 sin Nan")
print("=====================================")
print(test3_sin_nan)


```

```python
# Reemplazar NaN por "None" en los datos de prueba
test_questions_data.fillna("None", inplace=True)
test_data.fillna("None", inplace=True)

# Concatenar los conjuntos de datos de preguntas y pruebas
test_combined = pd.concat([test_questions_data, test_data], ignore_index=True)

# Unir los conjuntos de entrenamiento
train_combined = pd.concat([train1_sin_nan, train2_sin_nan], ignore_index=True)

# Mostrar los conjuntos de entrenamiento y prueba
print("Conjunto de entrenamiento:")
print(train_combined.head())
print("\nConjunto de prueba:")
print(test_combined.head())
```
